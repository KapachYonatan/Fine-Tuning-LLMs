## 2.1 — Load Tokenizers

```python
import json, pandas as pd

# Load pre-computed tokenizer data (generated once via extract_tokenizer_data.py)
with open("tokenizer_data.json", encoding="utf-8") as f:
    tdata = json.load(f)

MODEL_IDS = list(tdata.keys())
print(f"Loaded data for {len(tdata)} models:")
for mid in MODEL_IDS:
    d = tdata[mid]
    print(f"  OK  {mid:<50s}  class={d['tokenizer_class']}  vocab={d['vocab_size']}")
```

## 2.2 — Tokenizer Properties

```python
rows = []
for mid in MODEL_IDS:
    d = tdata[mid]
    rows.append({
        "model_id":             mid,
        "tokenizer_class":      d["tokenizer_class"],
        "vocab_size_base":      d["vocab_size"],
        "vocab_size_full":      d["vocab_size_full"],
        "boundary_strategy":    d["boundary_strategy"],
        "special_token_count":  d["special_token_count"],
        "special_token_examples": ", ".join(d["special_token_examples"]),
        "has_byte_tokens":      d["has_byte_tokens"],
    })

df_props = pd.DataFrame(rows)
pd.set_option("display.max_colwidth", 80)
df_props
```

## 2.3 — Average Tokens per Word

**Corpus:** ~1000-word Prague texts (`docs/prague_text_en.txt` and `docs/prague_text_he.txt`).

**Method:** `words` = whitespace-split tokens; `avg = len(encode(text, add_special_tokens=False)) / len(text.split())`.

**Assumption:** whitespace splitting is a reasonable word-count proxy for both English and Hebrew.
Hebrew words are whitespace-delimited (unlike, e.g., Chinese), so the same method applies directly.

```python
avg_rows = []
for mid in MODEL_IDS:
    d = tdata[mid]
    avg_rows.append({
        "model_id":                    mid,
        "avg_tokens_per_english_word": d["avg_tokens_per_english_word"],
        "avg_tokens_per_hebrew_word":  d["avg_tokens_per_hebrew_word"],
        "he_en_ratio":                 d["he_en_ratio"],
    })

df_avg = pd.DataFrame(avg_rows)
df_avg
```

## 2.4 — Tokenization Differences Example

English text: `../../docs/english_text_part2.txt` (World Chess Championship 2026 passage).

We inspect:
1. Total token count for the full passage — a coarse measure of overall efficiency.
2. Token-by-token breakdown of the two spans (`Gukesh Dommaraju`, `The World Chess Championship`), which are the primary point of divergence across tokenizers.


```python
CHESS_TEXT = open("../../docs/english_text_part2.txt", encoding="utf-8").read().strip()
print("Text:", CHESS_TEXT)
print(f"\nWord count: {len(CHESS_TEXT.split())}")

print("\n── Total tokens per model ──")
for mid in MODEL_IDS:
    n = tdata[mid]["total_chess_tokens"]
    short = mid.split("/")[-1]
    print(f"  {short:<35s} {n:3d} tokens")
```

```python
SPANS = ["Gukesh Dommaraju", "The World Chess Championship"]

for span in SPANS:
    print(f"\n── Span: '{span}' ──")
    groups = {}
    for mid in MODEL_IDS:
        toks = tuple(tdata[mid]["span_tokenizations"][span])
        groups.setdefault(toks, []).append(mid.split("/")[-1])
        short = mid.split("/")[-1]
        print(f"  {short:<35s} → {list(toks)}")
    print(f"  Unique tokenizations: {len(groups)}")
    print("  Agreement groups:")
    for toks, models in groups.items():
        print(f"    {list(toks)}  ← {models}")
```

## 2.5 — Write `tokenizers.csv`

```python
csv_rows = []
for mid in MODEL_IDS:
    p = df_props[df_props["model_id"] == mid].iloc[0]
    a = df_avg[df_avg["model_id"] == mid].iloc[0]
    csv_rows.append({
        "model_id":                      mid,
        "tokenizer_type":                p["tokenizer_class"],
        "vocab_size":                    p["vocab_size_full"],
        "special_tokens":                p["special_token_count"],
        "word_boundary_strategy":        p["boundary_strategy"],
        "byte_fallback_or_byte_level":   p["has_byte_tokens"],
        "avg_tokens_per_english_word":   a["avg_tokens_per_english_word"],
        "avg_tokens_per_hebrew_word":    a["avg_tokens_per_hebrew_word"],
    })

df_csv = pd.DataFrame(csv_rows)
df_csv.to_csv("../../tokenizers.csv", index=False)
print("Saved tokenizers.csv")
df_csv
```
