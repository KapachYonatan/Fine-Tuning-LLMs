## 1.3 — Verification: Load configs from HuggingFace

Verify the manually extracted data programmatically by loading each model's `config.json` via the HuggingFace hub (does not download weights).

```python
from transformers import AutoConfig

MODEL_IDS = [
    "meta-llama/Llama-3.1-8B-Instruct",
    "mistralai/Mistral-7B-Instruct-v0.3",
    "Qwen/Qwen2.5-7B-Instruct",
    "allenai/OLMo-2-1124-7B-Instruct",
    "ibm-granite/granite-3.3-8b-instruct",
    "deepseek-ai/DeepSeek-V3",
    "HuggingFaceTB/SmolLM2-1.7B-Instruct",
    "microsoft/Phi-4-mini-instruct",
    "tiiuae/Falcon3-7B-Instruct",
    "dicta-il/dictalm2.0-instruct",
]

configs = {}
for model_id in MODEL_IDS:
    try:
        cfg = AutoConfig.from_pretrained(model_id, trust_remote_code=True)
        configs[model_id] = cfg
        print(f"OK  {model_id}")
    except Exception as e:
        print(f"ERR {model_id}: {e}")
```

```python
FIELDS = [
    "hidden_size",
    "num_hidden_layers",
    "num_attention_heads",
    "num_key_value_heads",
    "intermediate_size",
    "hidden_act",
    "max_position_embeddings",
    "vocab_size",
    "rope_theta",
    "rope_scaling",
    "sliding_window",
    "tie_word_embeddings",
]

rows = []
for model_id, cfg in configs.items():
    row = {"model_id": model_id}
    for field in FIELDS:
        row[field] = getattr(cfg, field, "N/A")
    rows.append(row)

df_verify = pd.DataFrame(rows)
pd.set_option("display.max_colwidth", 60)
df_verify
```
